```python
ct_scans = sitk.GetArrayFromImage(sitk.ReadImage("DAMW-Mass2.mhd", sitk.sitkFloat32))


imagen_GATE_dosisDM = ct_scans[0]

fig = plt.figure(figsize=(10,10))
plt.imshow(imagen_GATE_dosisDM)
plt.colorbar()


plt.show()
```

```python
ct_scans = sitk.GetArrayFromImage(sitk.ReadImage("DAMWTar-Dose.mhd", sitk.sitkFloat32))


imagen_GATE_dosis = ct_scans[0]

fig = plt.figure(figsize=(10,10))
plt.imshow(imagen_GATE_dosis)
plt.title("Dosis ponderado por Masa")
plt.colorbar()


plt.show()
```

```python
from scipy.optimize import curve_fit
from scipy import asarray as ar,exp
x=np.linspace(0,30,50)

plt.plot(x,ct_scans[0][24],label="Datos")
plt.title("Energia por volumen(Gy)")
plt.xlabel("Distancia (cm)")
y=ct_scans[0][24]

mean = sum(x * y) / sum(y)
sigma = np.sqrt(sum(y * (x - mean)**2) / sum(y))

def Gauss(x, a, x0, sigma):
       return a * np.exp(-(x - x0)**2 / (2 * sigma**2))

popt,pcov = curve_fit(Gauss, x, y, p0=[max(y), mean, sigma])
plt.plot(x,Gauss(x,*popt),label='fit')
plt.legend()
print("El sigma es",sigma)
```
